## CSV Write And Read

```python
import pandas as pd
```

```python
dic = {"A": [1, 2, 3,10,11], "B": [4, 5, 6,12,13], "C": [7, 8, 9,14,15]}

df = pd.DataFrame(dic)
df
```

```python
df.to_csv("text.csv", index=False)
```

```python
var = pd.read_csv("text.csv")
var
```

```python
df.to_csv("test_1.csv",index=False,header=["num1","num2","num3"])
```
