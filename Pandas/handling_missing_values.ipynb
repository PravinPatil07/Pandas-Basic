## Handling Missing Values

```python
import pandas as pd
```

```python
pizza = pd.read_csv('pizzas.csv')
pizza
```

```python
pizza.isnull().sum()
```

### Dropna And Fillna

```python
pizza_copy = pizza.copy()
```

```python
pizza_copy
```

```python
pizza_copy.dropna() # drop rows with NaN values
```

```python
pizza_copy.dropna(axis=1) # drop columns with NaN
```

```python
pizza_copy.dropna(how="any") # drop rows with any NaN values
```

```python
pizza_copy.dropna(subset=["price", "size"]) # drop rows with NaN values in price or size
```

```python
pizza_copy.fillna(0) # fill NaN values with 0
```

```python
pizza_copy.fillna({"pizza_id":"bbq_ckn_s","pizza_type_id":"cali_ckn","price": 14, "size": "M"}) # fill NaN values with specific values for each column
```

```python
pizza_copy.ffill() # fill NaN values with the previous value
```

```python
pizza_copy.bfill() # fill NaN values with the next value
```
